```python
import astropy.constants as const
import astropy.units as u
import numpy as np
```

```python
def calculate_orbital_velocity(M, R):
  """
  Calculate the orbital velocity
  Input: M - mass, R = orbital radius
  Output: v - orbital velocity
  """
  v = np.sqrt(ac.G * M / R)
  return(v)
```

```python
Mass_buldge = 1.5e10 * u.solMass
Radius_buldge = np.arange(30) * u.kpc
V = np.sqrt(const.G * Mass_buldge / Radius_buldge)

#print(V.to(u.km/u.s))

import matplotlib.pyplot as plt# importing plotting package

plt.plot(Radius_buldge.to(u.kpc), V.to(u.km/u.s), ".-")#Plotting velosity vs.
plt.xlabel('buldge Radius (kpc)'.format(Radius_buldge.unit))# add x axis
plt.ylabel('buldge Velocity (km/s)'.format(V.unit))# add y axis
plt.title('buldge Velocity vs. Radius for buldge')
plt.show()
```

```python
print(V.to(u.km/u.s))
print(Radius_buldge.to(u.kpc))
print(Mass_buldge.to(u.solMass))
```

```python
M_disk = 1e11 * u.solMass # Define disk total mass
R_disk = 10 * u.kpc # Define the outer edge of disk radius
density_disk = M_disk / (np.pi * (R_disk**2)) # here the density is in the unit of solar mass per square kpc
print(density_disk.to(1e6 * u.solMass/u.kpc**2))
```

```python
def calculatingEnclosedMassForDisk(R, density=318 * 1e6 * u.solMass/u.kpc**2):
    if R < 10 * u.kpc:
        M = np.pi * (R**2) * density
    else:
        R = 10 * u.kpc # any radius larger than 10 kpc will be trucated at 10 kpc because of the extent of the disk component
        M = np.pi * (R**2) * density
    return(M)
```

```python
print(calculatingEnclosedMassForDisk(1 * u.kpc).to(1e6 * u.solMass), "at 1 kpc") # to convert to million solar mass to increase readability
print(calculatingEnclosedMassForDisk(5 * u.kpc).to(1e6 * u.solMass), "at 5 kpc")
```

```python
def calculatingEnclosedMassForMilkyWay(R, density_disk=318 * 1e6 * u.solMass/u.kpc**2, M_bulge = 1e10 * u.solMass):
    """
    Note that the halo mass is missing here, so that is what you will work on this Friday
    """
    M_disk = calculatingEnclosedMassForDisk(R, density=density_disk)
    M_total = M_disk + Mass_buldge
    return(M_total)
```

```python
# Add this line at the beginning of the file
import astropy.constants as ac

# redefine v_bulge_arr to avoid confusion
v_bulge_arr = calculate_orbital_velocity(Mass_buldge, Radius_buldge)


# Define arrays to be calculated
M_disk = np.zeros(len(Radius_buldge)) * u.solMass # Don't forget unit here
M_total = np.zeros(len(Radius_buldge)) * u.solMass

# Calculating enclosed masses for the disk and total component
for i in np.arange(len(Radius_buldge)):
    M_disk[i] = calculatingEnclosedMassForDisk(Radius_buldge[i])
    M_total[i] = calculatingEnclosedMassForMilkyWay(Radius_buldge[i])

# Calculating the resulting orbital velocity due to each component
v_disk_arr = calculate_orbital_velocity(M_disk, Radius_buldge)
v_bulge_disk_arr = calculate_orbital_velocity(M_total, Radius_buldge)
```

```python
M_halo = 1e12 * u.solMass
R_halo = 122.64 * u.kpc
Dencity_halo = M_halo / (np.pi * (R_halo**3)*4/3)
Ma_halo = Dencity_halo * (Radius_buldge**3) * 4/3
print(Ma_halo.to(u.solMass))

v_halo_arr = calculate_orbital_velocity(Ma_halo, R_halo)
print(v_halo_arr.to(u.km/u.s))
```

```python
# Plot them all
plt.plot(Radius_buldge.to(u.kpc), V.to(u.km/u.s), color="red", label="Bulge") # note that here we can add label to the rotation curve caused by the Bulge
plt.plot(Radius_buldge.to(u.kpc), v_disk_arr.to(u.km/u.s), color="blue", label="Disk") # plotting disk
plt.plot(Radius_buldge.to(u.kpc), v_bulge_disk_arr.to(u.km/u.s), color="black", label="Bulge + Disk") # plotting bulge + disk components
plt.plot(Radius_buldge.to(u.kpc), v_halo_arr.to(u.km/u.s), color="green", label="halo")
plt.xlabel('R (kpc)')
plt.ylabel('v (km/s)')
plt.legend() # to show the legend of a figure
plt.show()
```
